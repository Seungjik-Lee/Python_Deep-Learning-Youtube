# 영어성적, 학점, 지원한 대학원 수준에 따라서 합격할 확률 예측

```python
import tensorflow as tf      #tensorflow 라이브러리를 tf라는 이름으로 사용
import pandas as pd       #pandas 라이브러리를 pd라는 이름으로 사용
import numpy as np      #numpy 라이브러리를 np라는 이름으로 사용
```

## 데이터 준비

```python
data = pd.read_csv('gpascore.csv')
print(data)
```

## 데이터 전처리

```python
data.isnull().sum() #데이터의 빈칸 세기
```

```python
data = data.dropna() #빈칸인 데이터 행 삭제
```

```python
data.fillna(100) #빈칸에 100이라는 값을 넣음
```

```python
data['gpa'] #gpa라는 열의 내용이 출력됨
```

```python
data['gpa'].min() #gpa라는 열의 최소값 출력 ex)max, mean, count...
```

## 모델 만들기

```python
#모델 만들기, Sequential은 신경망 레이어를 쉽게 만들어주는 함수
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='tanh'), #2의 제곱승으로 표현하는것이 관습
    tf.keras.layers.Dense(128, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid'), #결과로 나오는 출력레이어 (확률 하나를 예측할 때 0~1) ex)0.9 = 90%, 0.2 = 20%
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## 모델 학습

```python
x_Data = []
```

```python
for i, rows in data.iterrows():   #iterrows : data라는 데이터프레임을 가로 한줄씩 출력
  x_Data.append([rows['gre'], rows['gpa'], rows['rank']])
```

```python
y_Data = data['admit'].values
```

```python
model.fit(np.array(x_Data), np.array(y_Data), epochs = 1000, verbose = 0)
model.fit(np.array(x_Data), np.array(y_Data), epochs = 10)
```

## 예측하기

```python
predict_values = model.predict(x_Data)
```
